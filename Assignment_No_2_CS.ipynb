```python
x= iris.sepallength
plt.hist(x)
plt.title('Histogram for sepal length')
plt.axvline(x.mean(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(x.median(),color ='g',linestyle ='dashed',linewidth =3)
plt.xlabel('sepal length')
plt.ylabel('frequency')
plt.show()
```
